```python
from sklearn.model_selection import train_test_split  # 划分训练集验证集
from keras.utils import to_categorical  #one-hot编码函数
from data import Caltech101  # 自定义数据模块
from model import VGG11, VGG13, VGG16C, VGG16D, VGG19  # 自定义模型
```

```python
x, y = Caltech101().gen_data()  # 获取全部数据
x_train, x_valid, y_train, y_valid = train_test_split(x, y, test_size=0.2, random_state=2019)  # 固定seed划分数据集
y_train = to_categorical(y_train)
y_valid = to_categorical(y_valid)
print(y_train.shape, y_valid.shape)
```

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[2])
plt.show()
```

```python
vgg11 = VGG11((128, 128, 3), 101)
vgg13 = VGG13((128, 128, 3), 101)
vgg16c = VGG16C((128, 128, 3), 101)
vgg16d = VGG16D((128, 128, 3), 101)
vgg19 = VGG19((128, 128, 3), 101)
```

```python
from keras.callbacks import EarlyStopping
callbacks = [
    EarlyStopping(monitor='val_loss', patience=10),
]
vgg11.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
vgg13.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
vgg16c.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
vgg16d.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
vgg19.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

# 训练VGG11

```python
his_vgg11 = vgg11.fit(x_train, y_train, validation_data=(x_valid, y_valid), epochs=50, batch_size=128, shuffle=False, verbose=True)  # 不允许shuffle，保证不同模型在同一个验证集上 
```

# 训练VGG13

```python
his_vgg13 = vgg13.fit(x_train, y_train, validation_data=(x_valid, y_valid), epochs=50, batch_size=128, shuffle=False, verbose=True) 
```

# 训练VGG16C

```python
his_vgg16c = vgg16c.fit(x_train, y_train, validation_data=(x_valid, y_valid), epochs=50, batch_size=128, shuffle=False, verbose=True)  
```

# 训练VGG16D

```python
his_vgg16d = vgg16d.fit(x_train, y_train, validation_data=(x_valid, y_valid), epochs=50, batch_size=128, shuffle=False, verbose=True)
```

# 训练VGG19

```python
his_vgg19 = vgg19.fit(x_train, y_train, validation_data=(x_valid, y_valid), epochs=50, batch_size=128, shuffle=False, verbose=True)
```

# 损失变化

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
plt.plot(np.arange(len(his_vgg11.history['loss'])), his_vgg11.history['loss'], label='vgg11 train loss')
plt.plot(np.arange(len(his_vgg13.history['loss'])), his_vgg13.history['loss'], label='vgg13 train loss')
plt.plot(np.arange(len(his_vgg16c.history['loss'])), his_vgg16c.history['loss'], label='vgg16c train loss')
plt.plot(np.arange(len(his_vgg16d.history['loss'])), his_vgg16d.history['loss'], label='vgg16d train loss')
plt.plot(np.arange(len(his_vgg19.history['loss'])), his_vgg19.history['loss'], label='vgg19 train loss')
plt.legend(loc=0)
plt.subplot(1, 2, 2)
plt.plot(np.arange(len(his_vgg11.history['val_loss'])), his_vgg11.history['val_loss'], label='vgg11 valid loss')
plt.plot(np.arange(len(his_vgg13.history['val_loss'])), his_vgg13.history['val_loss'], label='vgg13 valid loss')
plt.plot(np.arange(len(his_vgg16c.history['val_loss'])), his_vgg16c.history['val_loss'], label='vgg16c valid loss')
plt.plot(np.arange(len(his_vgg16d.history['val_loss'])), his_vgg16d.history['val_loss'], label='vgg16d valid loss')
plt.plot(np.arange(len(his_vgg19.history['val_loss'])), his_vgg19.history['val_loss'], label='vgg19 valid loss')
plt.legend(loc=0)
plt.savefig('loss.png')
```
